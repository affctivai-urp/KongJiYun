```python
import numpy as np
import scipy.io
import numpy as np
import pandas as pd
import random
```

# Data Loading

```python
valence_label = []
valence_data = []

for i in range(1, 33):
    if i<10:
        mat_file = scipy.io.loadmat(f"./data_preprocessed_matlab/s0{i}.mat")
        valence_label.append(list(mat_file['labels'][:, 0]))
        valence_data.append(mat_file['data'])
    else:
        mat_file = scipy.io.loadmat(f"./data_preprocessed_matlab/s{i}.mat")
        valence_label.append(list(mat_file['labels'][:, 0]))
        valence_data.append(mat_file['data'])
```

```python
valence_label = np.array(valence_label)
valence_data
```

# ,  

```python
from sklearn.preprocessing import Binarizer
binarizer = Binarizer(threshold=4.5, copy=False)
binarizer.transform(valence_label)
valence_label
```

# Train, Test 

```python
from sklearn.model_selection import train_test_split
#  
train_data, test_data, train_label, test_label = train_test_split(valence_data, valence_label, test_size=0.2)
print(train_label.shape, test_label.shape)
```
